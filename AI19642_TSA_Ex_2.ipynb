Implement programs for visualizing time series data


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/Electric_Production (1).csv", parse_dates=['DATE'], index_col='DATE')

# Display first few rows
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Production'], color='blue', label="Electricity Production")
plt.title("Electricity Production Over Time")
plt.xlabel("Year")
plt.ylabel("Production")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Production'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Electricity Production")
plt.xlabel("Production Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['Production'], color='lightcoral')
plt.title("Box Plot of Electricity Production")
plt.ylabel("Production")
plt.show()
```
